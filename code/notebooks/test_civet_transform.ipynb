```python
import glob
import subprocess
import os

def obj_to_gifti(obj_fin, clobber: bool = False):
    gii_fin = obj_fin.replace('.obj', '.surf.gii')
    
    if not os.path.exists(gii_fin) or clobber:
        print('Converting', obj_fin, 'to', gii_fin)
        subprocess.run(f'ConvertSurface -i_mni {obj_fin} -o_gii {gii_fin}', shell=True, executable='/bin/bash')

    return gii_fin

obj_files = glob.glob(f'civet/*/*/*obj')

#gii_files = [obj_to_gifti(obj, clobber=True) for obj in obj_files]
```

```python

from matplotlib_surface_plotting import plot_surf
import surfalign.utils as utils
import nibabel as nib
import numpy as np

def _plot(label_fin, surf_fin, title, filename='/tmp/tmp.png'):
    data = nib.load(label_fin).darrays[0].data

    coords, faces = utils.load_mesh_ext(surf_fin)

    plot_surf(
        coords,
        faces,
        data, 
        rotate=[90, 270], 
        title=title,  
        filename=filename,
        pvals=np.ones(coords.shape[0]),
        vmin=np.percentile(data, 2),
        vmax=np.percentile(data, 98),
        cmap='nipy_spectral',
    )
```
